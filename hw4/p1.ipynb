```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Figure 1

```python
p1_1_df = pd.read_csv("p1-1_summary.txt", sep="\s+", header=None, index_col=0, names=summary_column_names)

display(p1_1_df.head())
```

```python
age_years = p1_1_df["Age"]
log_luminosity = p1_1_df["Luminosity"] # Log10 L

plt.figure(figsize=(16, 4), dpi=100)
plt.plot(age_years, log_luminosity)
plt.xlabel("Age [years]", fontsize=16)
plt.ylabel("$\log_{10}$ Luminosity [${L_\odot}$]", fontsize=16)
plt.title("Luminosity vs Age", fontsize=16)
plt.tight_layout()
plt.savefig("p1-1_luminosity_vs_age.png", dpi=300)
plt.show()
```

## Figure 2

```python
log_radius = p1_1_df["Radius"] # Log10 R

plt.figure(figsize=(16, 4), dpi=100)
plt.plot(age_years, log_radius)
plt.xlabel("Age [years]", fontsize=16)
plt.ylabel("$\log_{10}$ Radius [${R_\odot}$]", fontsize=16)
plt.title("Radius vs Age", fontsize=16)
plt.tight_layout()
plt.savefig("p1-1_radius_vs_age.png", dpi=300)
plt.show()
```

## Figure 3

```python
log_temperature = p1_1_df["Surface temperature"] # Log10 T

plt.figure(figsize=(16, 4), dpi=100)
plt.plot(age_years, log_temperature)
plt.xlabel("Age [years]", fontsize=16)
plt.ylabel("$\log_{10}$ Surface temperature [K]", fontsize=16)
plt.title("Surface temperature vs Age", fontsize=16)
plt.tight_layout()
plt.savefig("p1-1_temperature_vs_age.png", dpi=300)
plt.show()
```

## Figure 4

```python
pp_chain_luminosity = p1_1_df["Luminosity from PP chain"]

plt.figure(figsize=(16, 4), dpi=100)
plt.plot(age_years, pp_chain_luminosity)
plt.xlabel("Age [years]", fontsize=16)
plt.ylabel("PP chain Luminosity [$L_\odot$]", fontsize=16)
plt.title("PP Chain Luminosity vs Age", fontsize=16)
plt.tight_layout()
plt.savefig("p1-1_pp_chain_luminosity_vs_age.png", dpi=300)
plt.show()
```

## Figure 5

```python
triple_alpha_luminosity = p1_1_df["Luminosity from triple-alpha reactions"]
cno_cycle_luminosity = p1_1_df["Luminosity from CNO cycle"]

plt.figure(figsize=(16, 4), dpi=100)
plt.plot(age_years, np.log10(pp_chain_luminosity), label="PP chain")
plt.plot(age_years, np.log10(triple_alpha_luminosity), label="Triple-alpha")
plt.plot(age_years, np.log10(cno_cycle_luminosity), label="CNO cycle")
plt.xlabel("Age [years]", fontsize=16)
plt.ylabel("$\log_{10}$ Luminosity [$L_\odot$]", fontsize=16)
plt.ylim(-6, 6)
plt.title("Luminosity vs Age", fontsize=16)
plt.legend()
plt.tight_layout()
plt.savefig("p1-1_luminosity_vs_age_components.png", dpi=300)
plt.show()
```
