我們現在探索一下產生能量的核融合反應，在 summary file 中有提供 pp chain，Triple-alpha reaction，CNO cycle，和 metal burning 的光度，請先畫出 x-axis 為 Age (years) 和 y-axis 為 Luminosity pp chain 的圖 [請好好標註單位和xy軸的名稱]，請附上你的圖(圖四)。[注意該資料沒有取Log10]

```python
pp_chain_luminosity = p1_1_df["Luminosity from PP chain"]

plt.figure(figsize=(8, 4), dpi=100)
plt.plot(age_years, pp_chain_luminosity)
plt.xlabel("Age (years)")
plt.ylabel("Luminosity from PP chain ($L_\odot$)")
plt.title("Luminosity from PP chain vs Age")
plt.show()
```

請問該恆星使用 pp chain 為主要能量來源到大約幾歲？後來的主要核融合反應為何？請畫圖說明(圖五)。[提示：將 pp chain、Triple-alpha reaction、CNO cycle 的光度 畫在同個圖上，再根據該圖判斷並說明。]

```python
triple_alpha_luminosity = p1_1_df["Luminosity from triple-alpha reactions"]
cno_cycle_luminosity = p1_1_df["Luminosity from CNO cycle"]

plt.figure(figsize=(10, 4), dpi=100)
plt.plot(age_years, np.log10(pp_chain_luminosity), label="PP chain")
plt.plot(age_years, np.log10(triple_alpha_luminosity), label="Triple-alpha")
plt.plot(age_years, np.log10(cno_cycle_luminosity), label="CNO cycle")
plt.xlabel("Age (years)")
plt.ylabel("Luminosity ($\log_{10}{L_\odot}$)")
plt.ylim(-6, 6)
plt.title("Luminosity vs Age")
plt.legend()
plt.show()
```

我們來做一個有趣的計算，上課說到 p-p chain 核融合反應會產生微中子，後來也有觀測到，請問微中子的光度 (L neutrino) 和 pp chain 所產生的光度 (Luminosity pp chain)，比例為何？(可以根據 summary file 中的資訊，計算 initial mass 為 1 恆星在主序帶上某個時期的比例即可)

```python
neutrino_luminosity = p1_1_df["Luminosity of neutrino losses"]

plt.figure(figsize=(8, 4), dpi=100)
plt.plot(age_years, np.log10(neutrino_luminosity/pp_chain_luminosity))

plt.xlabel("Age (years)")
plt.ylabel("Ratio ($\log_{10}{L_{ν}/L_{PP}}$)")
plt.title("Ratio of Neutrino Luminosity to PP chain Luminosity")
# plt.legend()
plt.show()
```
